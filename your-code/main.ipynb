#### Use your previous code but now, resize your plot.

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
plt.plot(x, x**2, linewidth=5)
plt.plot(x, np.exp(x), linewidth=5)

plt.xscale("log")
plt.yscale("log")
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# MEDIAN STEPS BY MONTHS_ENCODED
# your code here
ds.groupby("Months_encoded")["Steps"].median().plot(style="o-", ax=axs[0])
axs[0].set_xlim(0, 13)
axs[0].set_title("MEDIAN STEPS BY MONTHS_ENCODED")

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
ds.groupby("Work_or_Weekend")["Steps"].median().plot.bar(ax=axs[1])
axs[1].set_ylim(9900, 10500)
axs[1].set_title("MEDIAN STEPS BY WORK_OR_WEEKEND")

fig.savefig("figures/plots.jpg")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
cols = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]

for i,col in enumerate(cols):
    axs[i].scatter(ds["Steps"], ds[col], s=10)
    axs[i].set_title(col + " vs Steps")

fig.savefig("figures/plots2.jpg")
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
sns.boxplot(titanic["Age"]);
```

```python
"""
In the boxplot, it is possible to see:
* the minimum value - left-most line,
* the maximum value - right-most line,
* first quartile value - left edge of the shaded rectangle,
* second quartile value (=median) - line in middle of the rectangle,
* third quartile value - right edge of the rectangle, and
* outliers - dots to the right of the graph
"""
```

#### What does the last plot tell you about the feature `Age`?

```python
"""
Minimum value (excluding outliers): around 0 years of age
Q1: around 21
Median: around 28
Q3: around 37
Maximum value (excluding outliers): around 60
Outliers: present in the range 60-80
"""
```
